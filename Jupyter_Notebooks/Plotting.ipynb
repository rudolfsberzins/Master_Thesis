```python
from __future__ import division
import pandas as pd
from sys import argv
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
from matplotlib import style

style.use('fivethirtyeight')
```

```python
def open_model(model):
    
    model_df = pd.read_csv(model, sep = "\t", names = ["Classification Model", "Feature Count", "Error", "Time"])
    
    return model_df
```

```python
model_df = open_model("./Results/Full_Models.tsv")
```

```python
for i,r in model_df.iterrows():
    if r["Classification Model"] == "Random Forest (100 trees, random state = 23)":
        ran_for = plt.scatter(r["Feature Count"], r["Error"], s=70, marker = 'o', c = np.log(r["Time"]), label = r["Classification Model"])
    if r["Classification Model"] == "Neural Network":
        nn = plt.scatter(r["Feature Count"], r["Error"], s=70, marker = '^', c = np.log(r["Time"]), label = r["Classification Model"])
#     if r["Classification Model"] == "Linear Regression":
#         lin_reg = plt.scatter(r["Feature Count"], r["Error"], s=70, marker = 'D', c = np.log(r["Time"]), label = r["Classification Model"])
    if r["Classification Model"] == "Logistic Regression":
        log_reg = plt.scatter(r["Feature Count"], r["Error"], s=70, marker = 'v', c = np.log(r["Time"]), label = r["Classification Model"])
    if r["Classification Model"] == "Support Vector Machines":
        svm = plt.scatter(r["Feature Count"], r["Error"], s=70, marker = 's', c = np.log(r["Time"]), label = r["Classification Model"])
#     if r["Classification Model"] == "Gaussian Naive Bayes":
#         gnb = plt.scatter(r["Feature Count"], r["Error"], s=70, marker = 'x', c = np.log(r["Time"]), label = r["Classification Model"])
    if r["Classification Model"] == "K Nearest Neighbor (5)":
        knn = plt.scatter(r["Feature Count"], r["Error"], s=70, marker = '>', c = np.log(r["Time"]), label = r["Classification Model"])

plt.title("Classification Model Comparison")
plt.xlabel('Bag of Words Feature Count')
plt.ylabel('Error')
plt.colorbar()
# plt.legend(handles = [ran_for, nn, lin_reg, log_reg, svm, gnb, knn], loc = 4, prop={'size':4})
plt.show()
```
